### We are going to use the Pokemon data set to show the power of Pandas

The pokemon data set consists of 801 types of pokemon containing several features and attributes of each pokemon.

It is a nice and simple dataset to easily introduce you to using pandas

Unlike **Excel** we need to first define some libraries to import and use in order to be able to manipulate the data and plot it.

```python
# pandas is imported and aliased to 'pd'
import pandas as pd

# numpy is good at manipulating data and arrays
# pandas uses it heavily and it will come in handy
# we will import it and call it 'np'
import numpy as np

# for plotting we use matplotlib
import matplotlib.pyplot as plt
```

```python
# we can now read our data file
# the dataset is stored in Comma-Separated Values (CSV) format
# which is a text file with columns separated by commas
# an example would be:


# name,X,Y,Z
# bob,4,5.6,0.9e3
# jess,3,5.89,0.16e4
# ...

# read the data file
data = pd.read_csv('pokemon.csv')

# we could similarly read from Excel if it was in this format too!
# data = pd.read_excel('pokemon.xlsx')
```

```python
# we can now peak at our data using the head command
# this will show us the top 10 entries
data.head(10)
```

```python
# sometimes it is nice to see all the columns (not shown by default)
# so we can set this using the below command
pd.set_option('display.max_columns', None)
```

```python
# let's look again
data.head(10)
```

```python
# we can see the number of rows (entries) and the number of features (columns)
# the first value is the number of entries - should be 801
# the second value is the number of columns - should be 41
data.shape
```

```python
# we can already get a quick feel for the data by looking at some
# simple stats using describe
data.describe()
```

```python
# now it is always good practice to first find any missing values
# typically these will be 'NaN' - Not a number
# we cannot process these values for numerical analysis
# let's count these to see which columns contain problematic entries.
data.isna().sum(axis=0)
```

```python
# let's clean up these missing values
# one simple but naive way would be to remove these entries altogther
# but that would dramaitically reduce our dataset and therefore 
# loose information

# we have 4 columns containing NaN values
# height_m, weight_kg, percentage_male, and type2

# let's first look at the hieght and weight NaNs
# which pokemon are these?
missing_hieghts = data[pd.isna(data['height_m'])]['name']
missing_wieghts = data[pd.isna(data['weight_kg'])]['name']
missing_h_and_w = list(set(missing_hieghts) & set(missing_wieghts))
len(missing_h_and_w), len(missing_hieghts), len(missing_wieghts)
```

```python
# it seems to be the same 20 missing both height and weight
# these are the following pokemon
data[data['name'].isin(missing_h_and_w)][['name', 'height_m', 'weight_kg', 'hp']]
```

```python
# we can deal with this in a few ways.
# the simplist would be to remove these entries
# since it is only 20 out of 801 (2.5%) it would not impact
# any analysis or predictions that much
# if we want to use these features to either make a prediction
# or use these in another fashion i.e. regression problem
# predict the height of a pokemon based on X,Y,Z...
# then it is important to have accurate values

# For now we will simply drop these 20 (remove them)
processed_data = data[~data['name'].isin(missing_h_and_w)]

# Further extension:
# It is possible to input these correct values which can easily be found online
# there is an API to do this https://pokeapi.co/
```

```python
# check for NaNs again
processed_data.isna().sum(axis=0)
```

```python
plt.hist(processed_data['percentage_male'], bins=range(0, 100, 1))
average_pct_male = np.median(processed_data['percentage_male'].dropna())
average_pct_male
```

```python
# for the case of percentage male
# we do not drop them but simply replace it with 50% (median value)
processed_data['percentage_male'] = processed_data['percentage_male'].fillna(average_pct_male)
processed_data.isna().sum()
```

```python
# now we just have to deal with the NaNs in type2
# which have the largest number - 384

# we will have a look at the unique entries for this type
processed_data['type2'].unique()
```

```python
# we notice that type2 and type1 have similar groups
processed_data['type1'].unique()
```

```python
# let's check they are the same - just to be sure
# we can see there are 18 types for a pokemon, with each have either one or two types
types = list(set(processed_data['type1'].unique()) & set(processed_data['type2'].unique()))
len(types), processed_data['type1'].nunique(), processed_data['type2'].nunique()
```

```python
# it would actually be better to turn this into 18 columns instead of 2 with 1 indicating the
# types, and 0 for all else
# we can do this with get_dummies
final_data = pd.get_dummies(processed_data, columns=['type1', 'type2'], prefix=['type','type'])
# we then need to merge columns of the same name
final_data = final_data.groupby(level=0, axis=1).sum()

# check for NaNs again
final_data.isna().sum(axis=0)
```

```python
# now we have no more NaNs and we only lost 20 pokemon!
final_data.shape
```

```python
# let's remove the japanese name, we don't need this
cols_to_remove = [
    'japanese_name'
]

final_data.drop(columns=cols_to_remove, inplace=True)
final_data
```

```python
# we still have a few trouble some columns
# that is columns containing non numerical data (strings)
# these are:
#   'abilities', 'classfication', 'name'

# first let's check that name is unique
assert final_data['name'].nunique() == final_data.shape[0]
# and let's check that pokedex_number is unique too
assert final_data['pokedex_number'].nunique() == final_data.shape[0]

# since name is unique we can use this to idenfity each pokemon

# let's look at the classification column - which is incorrectly spelt
# let's rename it first
final_data.rename(columns={'classfication': 'classification'}, inplace=True)
final_data['classification'].nunique()

# since this is a large number of unique categories we will drop this column
final_data.drop(columns=['classification'], inplace=True)
final_data.shape
```

```python
# that just leaves abilities which is a bit more tricky to handle 
# since it is an array of strings

# we need to break the array up and again count the number of abilities
# again, if this is to large to categorise we will ignore it for now
# but note that we can do other interesting things with this using some 
# natural language processing techniques (not covered here)

abilities = final_data['abilities'].values

# need to convert from an array of strings representing arrays
# to a full array of abilities
all_abilities = []
for a in abilities:
    entries = eval(a)
    all_abilities.extend(entries)

# 223 is too large to categorise, let's drop abilities for now
len(set(all_abilities))
```

```python
final_data.drop(columns=['abilities'], inplace=True)
final_data.shape
```
